```python
df = pd.read_csv("data/shop_smart_ecommerce.csv")
```

```python
## Checking the null for all cols

df.isnull().sum()
```

```python
### Number and catergrical cols separte ---> Using df.select_dtypes 

categorical_cols = df.select_dtypes(include = ["object"]).columns
numerical_cols = df.select_dtypes(include = ["number"]).columns
boolen_cols = df.select_dtypes(include= ["bool"]).columns
```

```python
## EDA on Target to know what is happening
# Return cnt of false(Not buy) & true(Buy) 
#value_count() works on any dtype regardless

df['Revenue'].value_counts()
df['Revenue'].value_counts(normalize=True)  # class balance — likely ~85% False, 15% True
```

```python
# Directly gives True(buy %)
df['Revenue'].mean()   # 0.155 → same as the True fraction above
```

```python
#Target Distribution
#False - Not buy  True- Buy
# Only 2k are purchasing and 10k not buying

sns.countplot(data=df, x='Revenue')
plt.title('Purchase vs No Purchase')
plt.show()
```

```python
# ## Numeric features vs Revenue — boxplots show which behaviors differ 
# # between buyers and non-buyers:

# num_cols = ['Administrative_Duration', 'Informational_Duration',
#             'ProductRelated_Duration', 'BounceRates', 'ExitRates', 'PageValues']
# # numerical_cols

# fig, axes = plt.subplots(2, 3, figsize=(16, 8))
# for ax, col in zip(axes.flatten(), num_cols):
#     sns.boxplot(data=df, x='Revenue', y=col, ax=ax)
#     ax.set_title(col)
# plt.tight_layout()
# plt.show()

sns.boxplot(data=df, x="Revenue", y="ExitRates")
```

```python
#Correlation heatmap

plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Categorical features 

for col in ['Month', 'VisitorType', 'Weekend']:
    conv = df.groupby(col)['Revenue'].mean().sort_values(ascending=False)
    conv.plot(kind='bar', figsize=(8, 4), title=f'Conversion rate by {col}')
    plt.ylabel('Conversion rate')
    plt.show()
```
